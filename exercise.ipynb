```python
movies = [
    {
        "title": "The Shawshank Redemption",
        "director": "Frank Darabont",
        "year": 1994,
        "genre": ["Drama"]
    },
    {
        "title": "The Godfather",
        "director": "Francis Ford Coppola",
        "year": 1972,
        "genre": ["Crime", "Drama"]
    }
]
```

```python
#add new movie to movies list
new_movie = {
    "title": "Inception",
    "director": "Christopher Nolan",
    "year": 2010,
    "genre": ["Action", "Sci-Fi"]
}
movies.append(new_movie)
for movie in movies:
    print(f"Title: {movie['title']}, Director: {movie['director']}, Year: {movie['year']}, Genre: {', '.join(movie['genre'])}")
```

```python
#search movie by title or dierector
search_title = "Inception"
search_director = "Christopher Nolan"
for movie in movies:
    if movie['title'] == search_title or movie['director'] == search_director:
        print(f"Found Movie - Title: {movie['title']}, Director: {movie['director']}, Year: {movie['year']}, Genre: {', '.join(movie['genre'])}")
```
